```python
import pandas as pd
import pandas.io.parsers
import pprint
import matplotlib.pyplot as plt
import matplotlib
```

```python
equity_file = 'equity.csv'

equity_data = pandas.io.parsers.read_csv(equity_file, header=1, delimiter=',', index_col=0, \
                                         names=['datetime', 'cash','commission','eurusd','total','returns','equity_curve','drawdown'], \
                                         parse_dates=True
)
```

```python
%matplotlib inline
matplotlib.style.use('ggplot')
```

```python
 equity_data.plot(y='equity_curve')
```

```python
 equity_data.plot(y='drawdown')
```

```python
 equity_data.plot(y='total')
```
